By using Yelp's API, the McDonald's located in London has been imported to the notebook in put into an empty list.
As there are dictionary from Yelp consists of different restaurant, a filter has been put in the list to ensure all data extracted are McDonald's.

```python
zipcodes = []

for business in allBusinesses:
    if("mcdonalds" in business['name'].replace("'", "").lower()):
        zipcodes.append(business['location']['zip_code'])

print(len(zipcodes))       
print(zipcodes)
```

```python
coordinates = []
for business in allBusinesses:
    coordinates.append(business['coordinates'])
    
print (coordinates)
```

London population has been extracted from information provided by the government website.

```python
import pandas as pd
london_population = pd.read_excel("london-borough-profiles.xlsx", 'Data')
```

```python
borough = london_population['Area_name'].values
print (borough)
```

```python
population = london_population['GLA Population Estimate 2017'].values
print (population)
```

```python
borough_population = dict(zip(borough, population))
print (borough_population)
```

The data from different column has been combined into a dictionary to match with the McDonald's postcode.

```python
del borough_population['Inner London'] 
del borough_population['Outer London'] 
del borough_population['London'] 
del borough_population['England']
del borough_population['United Kingdom']
del borough_population['National comparator']
print(borough_population)
```

```python
london_postcode = pd.read_excel("London postcodes.xlsx")
```

```python
mcd_match = london_postcode.loc[london_postcode['Postcode'].isin(zipcodes)]
mcd_match
```

```python
mcd_in_london = dict(mcd_match.groupby(['District'])['Postcode'].count().sort_values())
mcd_in_london
```

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go

plotly.offline.init_notebook_mode(connected=True)
```

```python
trace1 = go.Bar(
    x = list(borough_population.keys()),
    y = list(borough_population.values()),
    name='London Population'
)
trace2 = go.Bar(
    x = list(mcd_in_london.keys()),
    y = list(mcd_in_london.values()),
    name='Number of MCD in London',
    yaxis='y2'
)

data = [trace2, trace1]

layout = go.Layout(
    title='London Population & Number of MCD in each borough',
    yaxis=dict(
        title='Population'
    ),
    yaxis2=dict(
        title='Number of MCD',
        titlefont=dict(
            color='rgb(148, 103, 189)'
        ),
        tickfont=dict(
            color='rgb(148, 103, 189)'
        ),
        overlaying='y',
        side='right'
    )
)
fig = go.Figure(data=data, layout=layout)
# py.iplot(fig)
plotly.offline.iplot(fig)
```

Looking at the graph, there's little correlation between the population of each borough and the number of McDonald's in one borough. For example, Westminister has the most number of McDonald's as one of the highest population amongst all boroughs, while Barnet and Cryodon also are two the highest populations but having 4x less McDonald's in the district.
This is not taken account that each borough might spread across different zones, which can also attribute to the traffic and population of that borough. 
Taking out Westminster as a outlier of the dataset, the population is still not a strong variable attributing to the number of McDonald's in a borough.
In conclusion, the population of a borough does not have any impact on the number of McDonald's of one borough.
